# **Numpy Library - Numerical Python**

Advantages of Numpy Arrays

1. Allows several Mathematical Operations
2. Faster Operations

```python
import numpy as np
```

List vs Numpy - Time Taken

```python
from time import process_time
```

Time taken by list

```python
python_list = [i for i in range(10000)]

start_time = process_time()
python_list = [i+5 for i in python_list]
end_time = process_time()
print(end_time - start_time)
```

Time taken by Numpy array

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()
np_array += 5
end_time = process_time()
print(end_time - start_time)
```

Numpy Arrays

```python
# list
list = [1,2,3,4,5]
print(list)
type(list)
```

```python
# numpy array
numpy_array = [1,2,3,4,5]
print(numpy_array)
type(numpy_array)
```

```python
# 1 dimensional array
a = np.array([1,3,4,5,3])
print(a.shape)

b = np.array([(2,43,5,2,2,5,2),(2,24,3,6,4,2,7)])
print(b.shape)

c = np.array([(2,43,5,2,2,5,2),(2,24,3,6,4,2,7)], dtype=float)
print(c)
```

Initial Placeholders in Numpy Array

```python
# array of zeros
nparray1 = np.zeros((4,5))
print(nparray1)

# array of ones
nparray2 = np.ones((3,3))
print(nparray2)

# array of particular value
nparray3 = np.full((3,3),100)
print(nparray3)
```

```python
# Identity matrix = diagonal values are one
nparray4 = np.eye(5)
print(nparray4)
```

```python
# create a numpy with random values
npram1 = np.random.random((3,4))
print(npram1)

# create a numpy with random values within specific range
npram2 = np.random.randint(10,100,(3,5))
print(npram2)
```

```python
# array of an evenly spaced values --> specifying the number of values required
nparray5 = np.linspace(10,30,6)
print(nparray5)
```

```python
# array of an evenly spaced values --> specifying the step
nparray6 = np.arange(10,30,5)
print(nparray6)
```

```python
# list to a numpy array
# similarly tuple can be converted to numpy array
list2 = [3,5,2,2,4]
nparray7 = np.asarray(list2)

print(nparray7)
type(nparray7)
```

# Mathematical Operations on an Numpy Array

```python
list1 = [2,5,2,6,7]
list2 = [3,5,6,2,2]
print(list1 + list2)
```

```python
np1 = np.random.randint(10,30,(3,3))
np2 = np.random.randint(30,60,(3,3))

print(np1)
print(np2)
```

```python
print(np1+np2)
print(np1-np2)
print(np1*np2)
print(np1/np2)

print(np.add(np1,np2))
print(np.subtract(np1,np2))
print(np.multiply(np1,np2))
print(np.divide(np1,np2))
```

# Array Manipulation

```python
arrman = np.random.randint(0,10,(2,3))
print(arrman)
print(arrman.shape)
```

```python
# transpose
trans = np.transpose(arrman)
print(trans)
print(trans.shape)

# another method of transpose
trans2 = arrman.T
print(trans2)
print(trans2.shape)
```

```python
# reshaping array
a = np.random.randint(0,10,(2,3))
b = a.reshape(3,2)
print(a)
print(a.shape)
print(b)
print(b.shape)
```
